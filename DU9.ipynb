```python
import pandas
```

```python
tmprtrs=pandas.read_csv('temperature.txt')
tmprtrs.head()
```

Dotaz na měření, která byla provedena v Praze. Je na datech něco zvláštního? Napadá tě, čím to může být? Zde je nápověda.

```python
tmprtrs[tmprtrs['City']=='Prague']
# Vypada to na stupne Fahrenheita
# Prepocet do stupnu Celsia
tmprtrs['avg_C']=5/9 * (tmprtrs['AvgTemperature']-32)
tmprtrs[tmprtrs['City']=='Prague']
```

Dotaz na měření, ve kterých je teplota (sloupec AvgTemperature) vyšší než 80 stupňů.

```python
tmprtrs[tmprtrs['AvgTemperature']>80]
```

Dotaz na měření, ve kterých je teplota vyšší než 60 stupňů a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).

```python
tmprtrs[(tmprtrs['AvgTemperature']>60) & (tmprtrs['Region']=='Europe')]
```

Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 80 stupňů nebo menší než -20 stupňů.

```python
tmprtrs[(tmprtrs['AvgTemperature']>80)  | (tmprtrs['AvgTemperature']<-20) ]
# kontrola
tmprtrs[(tmprtrs['AvgTemperature']>80)  | (tmprtrs['AvgTemperature']<-20) ].to_csv(path_or_buf='Extremes.csv', sep=',')
```
